```python
# 9 Advance Booking Trend: On average, how many days in advance do passengers book their tickets? (Compare Journey Date with Booked On)
# Convert to datetime
df['Journey Date'] = pd.to_datetime(df['Journey Date'])
df['Booked On'] = pd.to_datetime(df['Booked On'])

# Create new column for Advance Days
df['Advance Days'] = (df['Journey Date'] - df['Booked On']).dt.days

# Average Advance Booking Days
avg_advance_days = df['Advance Days'].mean()

print("Average Advance Booking Days:", round(avg_advance_days, 2))
```

```python
# 12 OTA vs. Local: Compare the total revenue generated by "OTA Agent Bookings" vs "Online Agent Bookings".
df['Booked By'] = df['Booked By'].astype(str).str.strip().str.lower()
# OTA Revenue
ota_revenue = df[df['Booked By'] == 'ota agent']['Net Amount'].sum()

# Online Agent Revenue
online_revenue = df[df['Booked By'] == 'online agent']['Net Amount'].sum()

print("OTA Agent Revenue:", ota_revenue)
print("Online Agent Revenue:", online_revenue)
```

```python
# 13 Route Profitability: Identify routes where the Service Status is frequently "Stopped." Analyze if these routes also have the highest cancellation rates or low net revenue.
# Clean text columns (important)
df['Service Status'] = df['Service Status'].astype(str).str.strip().str.lower()
df['Status'] = df['Status'].astype(str).str.strip().str.lower()

# Create Cancel Flag
df['Cancel Flag'] = df['Status'].apply(lambda x: 1 if x == 'cancel' else 0)

# Group by Route
route_analysis = df.groupby('Route').agg(
    total_revenue = ('Net Amount','sum'),
    stop_count = ('Service Status', lambda x: (x == 'stopped').sum()),
    total_bookings = ('Service Status','count'),
    cancel_rate = ('Cancel Flag','mean')
)

# Convert cancel rate to %
route_analysis['cancel_rate'] = route_analysis['cancel_rate'] * 100

# Sort by stop_count (frequent stopped routes on top)
route_analysis = route_analysis.sort_values(by='stop_count', ascending=False)

print(route_analysis.head(10))
```

```python
top_routes = route_analysis.head(10)

plt.figure()
plt.scatter(
    top_routes['cancel_rate'],
    top_routes['total_revenue'],
    s=top_routes['stop_count']*20
)

for i, route in enumerate(top_routes.index):
    plt.text(
        top_routes['cancel_rate'].iloc[i],
        top_routes['total_revenue'].iloc[i],
        route
    )

plt.xlabel("Cancel Rate (%)")
plt.ylabel("Total Revenue")
plt.title("Route Profitability vs Stops Analysis")

plt.show()
```

```python
#17.Churn Risk: If a customer has multiple "Cancel" entries in a row, how would you identify them as a "high-risk churn" passenger using this dataset, and what strategy would you suggest to retain them?
```

```python
df['Name'] = (
    df['Name']
      .astype(str)
      .str.strip()      
      .str.lower()      
)
```

```python
# Sort by customer and booking date
df = df.sort_values(['Name','Booked On'])

# Create Cancel Flag
df['Cancel Flag'] = df['Status'].str.lower().eq('cancel').astype(int)

# Count consecutive cancels per customer
df['Consecutive Cancels'] = df.groupby('Name')['Cancel Flag'] \
                               .rolling(window=3, min_periods=3) \
                               .sum().reset_index(level=0, drop=True)

# High-Risk Customers (3 cancels in a row)
high_risk_customers = df[df['Consecutive Cancels'] == 3]['Name'].unique()

print("High Risk Churn Customers:")
print(high_risk_customers)
```
